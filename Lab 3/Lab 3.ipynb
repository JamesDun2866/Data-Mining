```python
import pandas as pd #importing
```

```python
train = pd.read_csv("train.csv").copy() # Importing csv
```

```python
Embarked = train['Embarked'] # setting ID to a varible [[]] is a Dataframe [] is a series
```

```python
print(Embarked.unique()) #prints the unique varibles 
```

```python
Embarked.value_counts() #amount of each value
```

```python
nullcheck = Embarked.isnull() #converts to true/false values if a value is null
count =0 # sets count to zero (used to count nulls)
countmax = Embarked.count() # max amount in data

for x in nullcheck:#loop around all values
    if x == True:#if null is true
        count = count +1#add to a nullcount
    
percentage = count/countmax*100 #work out percentage of nulls
print ("the percentage of nulls is",percentage)
```
