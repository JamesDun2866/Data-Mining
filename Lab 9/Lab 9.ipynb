```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import MeanShift
from sklearn.cluster import DBSCAN
```

```python
#SimpleCluster = pd.read_csv("simple_cluster.csv").copy() # Importing csv
SimpleCluster = pd.read_csv("spiral_data.csv").copy() # Importing csv
SimpleClusterCopy = SimpleCluster.copy() #make a copy of the data

SimpleClusterCopy.head() #Display the top 5 records
```

```python
sns.scatterplot(x= SimpleClusterCopy['x'], y = SimpleClusterCopy['y'])
```

```python
kmeans = KMeans(n_clusters=2,init ='random').fit(SimpleClusterCopy)
```

```python
kmeanscentre = kmeans.cluster_centers_
kmeanscentre
```

```python
sns.scatterplot(SimpleClusterCopy['x'],SimpleClusterCopy['y'])
sns.scatterplot(kmeanscentre[:, 0], kmeanscentre[:, 1], s=200);
```

```python
sns.scatterplot(SimpleClusterCopy['x'],SimpleClusterCopy['y'],hue=kmeans.predict(SimpleClusterCopy) )
sns.scatterplot(kmeanscentre[:, 0], kmeanscentre[:, 1], s=200);
```

```python
meanshift = MeanShift().fit(SimpleClusterCopy)
```

```python
meanshiftcentre = meanshift.cluster_centers_
meanshiftcentre
```

```python
sns.scatterplot(SimpleClusterCopy['x'],SimpleClusterCopy['y'])
sns.scatterplot(meanshiftcentre[:, 0], meanshiftcentre[:, 1], s=200);
```

```python
sns.scatterplot(SimpleClusterCopy['x'],SimpleClusterCopy['y'],hue=meanshift.predict(SimpleClusterCopy) )
sns.scatterplot(meanshiftcentre[:, 0], meanshiftcentre[:, 1], s=200);
```

```python
dbscan = DBSCAN().fit(SimpleClusterCopy)
```

```python
sns.scatterplot(SimpleClusterCopy['x'],SimpleClusterCopy['y'],hue=dbscan.fit_predict(SimpleClusterCopy) )
```
