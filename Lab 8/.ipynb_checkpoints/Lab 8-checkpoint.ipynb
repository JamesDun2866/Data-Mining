```python
import pandas as pd
import seaborn as sns
import numpy as np
import sklearn as sk
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.utils.multiclass import unique_labels
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error 
from sklearn.metrics import mean_squared_error 
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn import tree as tree
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
```

```python
train = pd.read_csv("train.csv").copy() # Importing csv
traincopy = train.copy()
```

```python
traincopy = traincopy.drop(columns=['SibSp','Parch','Ticket','Cabin','PassengerId','Name'])
```

```python
traincopy
```

```python
print("Age Mean",traincopy['Age'].mean())
print("Mode of values:",'\n',traincopy.mode())
```

```python
traincopy[traincopy['Age'].isnull()]=traincopy[traincopy['Age'].isnull()].fillna('29')
traincopy[traincopy['Embarked'].isnull()]=traincopy[traincopy['Embarked'].isnull()].fillna('S')
```

```python
traincopy[['Fare']] = traincopy[['Fare']].values.astype(float)
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
traincopy[['Fare']] = min_max_scaler.fit_transform(traincopy[['Fare']])
```

```python
traincopy
```

```python
traincopy[['Age']] = traincopy[['Age']].values.astype(float)
```

```python
traincopy[['Age']] = min_max_scaler.fit_transform(traincopy[['Age']])
```

```python
traincopy
```

```python
dummiesSex = traincopy["Sex"].str.get_dummies(" ") #get dummies of clean data
dummiesSex.head() # show dummies
```

```python
dummiesEmbarked = traincopy["Embarked"].str.get_dummies(" ") #get dummies of clean data
dummiesEmbarked.head() # show dummies
```

```python
traincopy = traincopy.drop(columns=['Embarked','Sex'])
```

```python
frames = [traincopy,dummiesSex] #data to concatanate
traincopy = pd.concat(frames, axis=1) #concated data
```

```python
frames = [traincopy,dummiesEmbarked] #data to concatanate
traincopy = pd.concat(frames, axis=1) #concated data
```

```python
traincopy
```

```python
y = traincopy["Survived"]
y
```

```python
x = traincopy.drop(columns=['Survived'])
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.30, shuffle = True)
```

```python
Model = MLPClassifier(solver ='sgd',activation='logistic',max_iter=500000,hidden_layer_sizes =4505)
```

```python
Model.fit(X_train,y_train)
```

```python
yPredict = Model.predict(X_test)
yPredict
```

```python
#Code taken and editied from here(https://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html) added to make confusion matrixes look better and clearer than the standard array format


classes = ["0","1"] #set clasess for data analysing 
def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    if not title: #sets title of confustion matrix if there isnt one
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # set confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    # Only use the labels that appear in the data bhased on the set class data
    classes = classes
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

```python
plot_confusion_matrix(y_test, yPredict, classes=classes, title='Confusion matrix, without normalization')
```

```python
plot_confusion_matrix(y_test, yPredict, classes=classes, normalize=True, title='Normalized confusion matrix')
```

```python
accuracy= accuracy_score(y_test,yPredict,normalize = True)
print("Accuracy of Neural Network","%0.2f"%accuracy)
print("First Accuracy 66%")
print("Normalised Fare Accuracy 80%")
print("Normalised Age Accuracy 80%")
print("Arrucay of max Hidden layer 60%")
```
